# Current Conditions and Forecast

```python
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
time = np.arange(72)
wind = np.random.uniform(0,1,len(time))
data_dict = {'time': time, 'wind': wind}
df = pd.DataFrame(data_dict)
```

Wind Conditions

```python
fig = px.line(df, x="time", y="wind", title='Wind Conditions')
fig.show()
```
